```python
# 왜도(skew) : 어느정도로 비대칭인지?
# - 음수인 경우 : 왼쪽이 긴 꼬리를 갖는 형태(평균과 중위수 <최빈수)
# - 양수인 경우 : 오른쪽이 ... (평균과 중위수 > 최빈수)
# - 0인 경우 : 평균과 중앙값이 같은 경우(완전 대칭)
# -0.5~0.5 : 대체적으로 대칭적이다
# -1~-0.5, 0.5~1 : 어느정도 치우쳐 있다
# -1보다 작거나 1보다 큰 경우 : 많이 치우쳐 있다

df.skew()
```

```python
#첨도 : 확률분포의 뽀족한 정도, 특이값
#표준정규분포의 첨도가 3

#첨도가 3보다 작다 -> 특이값이 거의 없다
#첨도가 3보다 크다 -> 특이값이 있다
```
