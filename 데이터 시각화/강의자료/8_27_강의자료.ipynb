```python
#상관계수값이 0.7이상인 두 변수의 쌍을 모두 출력 -> 내림차순
corr
```

```python
li_co=list(corr.columns)
result = []
for i in range(len(corr)):
    for j in li_co:
            if abs(corr.iloc[i][j]) > 0.7:
                result.append([corr.iloc[i][j],corr.index[i],j])
sorted(result, reverse=True)
```

```python
for i in corr.keys():
    for j in corr.index:
        if corr.loc[j,i] > 0.7 and i != j:
            print((i,j))
```

```python
corr_s = corr.stack()
corr_s[corr_s > 0.7].sort_values(ascending=False).index
```
