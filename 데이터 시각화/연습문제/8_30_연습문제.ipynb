### 2.1 독립변수 전체(종류를 가진)와 SalePrice의 boxplot

```python
# 내부 값의 타입이 str인 경우 종류를 가진 독립변수로 판단
for col in df:
    if type(df.iloc[0][col]) is str:
        number_of_kinds = len(df[col].unique())
        if number_of_kinds > 8: # 종류가 너무 많으면 x축이 제대로 안보여서 크기를 바꿨음
            sns.catplot(x=col, y='SalePrice', kind='box', data=df, height=number_of_kinds)
        else:
            sns.catplot(x=col, y='SalePrice', kind='box', data=df, height=8)
```

## 3. 상관계수

```python
corr = df.corr()
```

```python
mask = np.triu(np.ones_like(corr,bool))
plt.figure(figsize = (28,20))
sns.heatmap(corr, annot = True, mask = mask, cmap = 'coolwarm', vmin = -1, vmax = 1)
```
